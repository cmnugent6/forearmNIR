```python
from PIL import Image
import numpy as np
import os

# define the paths to the folders containing the images
folder1_path = "/content/drive/MyDrive/biomed/firstRun/Images Used"
folder2_path = "/content/drive/MyDrive/biomed/firstRun/Makssjpg"

# get a list of all the files in each folder
folder1_files = os.listdir(folder1_path)
folder2_files = os.listdir(folder2_path)

# sort the files in each folder alphabetically
folder1_files.sort()
folder2_files.sort()

# iterate over the files and display them side by side
for i in range(len(folder1_files)):
    # get the file names of the images being displayed
    img1_name = folder1_files[i]
    img2_name = folder2_files[i]

    # open the images and convert them to grayscale
    img1 = Image.open(os.path.join(folder1_path, img1_name)).convert('L')
    img2 = Image.open(os.path.join(folder2_path, img2_name)).convert('L')

    # resize the images to the same size
    size = (200, 200)
    img1 = img1.resize(size)
    img2 = img2.resize(size)

    # convert the images to numpy arrays
    np_img1 = np.array(img1)
    np_img2 = np.array(img2)

    # create a new numpy array with the two images side by side
    combined_img = np.concatenate((np_img1, np_img2), axis=1)

    # create a new image object with the combined image
    combined_img = Image.fromarray(combined_img)

    # display the combined image and the file names
    print(f"{img1_name}\t{img2_name}")
    combined_img.show()
```

```python
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_dir = '/content/drive/MyDrive/biomed/firstRun/train'
val_dir = '/content/drive/MyDrive/biomed/firstRun/validation'
img_size = (256, 256)
batch_size = 32

# Define the ImageDataGenerator for data augmentation and preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

# Create the ImageDataGenerator flow from directory for training and validation sets
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='input',
    classes=['images', 'masks'],
    color_mode='rgb',
    interpolation='nearest'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='input',
    classes=['images', 'masks'],
    color_mode='rgb',
    interpolation='nearest'
)

# Print the input shape
print(f"Input shape: {train_generator[0][0].shape}")

# Define the model architecture
model = keras.models.Sequential([
    # encoder
    keras.layers.Conv2D(32, 3, activation='relu', padding='same', input_shape=(256, 256, 3)),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2D(32, 3, activation='relu', padding='same'),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling2D(pool_size=(2, 2)),
    keras.layers.Dropout(0.25),

    # decoder
    keras.layers.Conv2DTranspose(32, 3, strides=2, activation='relu', padding='same'),
    keras.layers.Conv2D(32, 3, activation='relu', padding='same'),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2DTranspose(3, 3, strides=2, activation='sigmoid', padding='same')
])

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=10,
    validation_data=val_generator,
    validation_steps=len(val_generator)
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set the directories for the original images and masks
train_image_dir = '/content/drive/MyDrive/biomed/firstRun/train_image'
train_mask_dir = '/content/drive/MyDrive/biomed/firstRun/train_mask'

# Define the image data generator for data augmentation
image_datagen = ImageDataGenerator(rotation_range=10,
                                   width_shift_range=0.1,
                                   height_shift_range=0.1,
                                   shear_range=0.1,
                                   zoom_range=0.1,
                                   horizontal_flip=True,
                                   fill_mode='nearest')

# Load the training data and masks using the same batch size
batch_size = 5
train_image_generator = image_datagen.flow_from_directory(train_image_dir,
                                                          target_size=(256, 256),
                                                          batch_size=batch_size,
                                                          class_mode=None,
                                                          seed=42)
train_mask_generator = image_datagen.flow_from_directory(train_mask_dir,
                                                         target_size=(256, 256),
                                                         batch_size=batch_size,
                                                         class_mode=None,
                                                         seed=42)

# Combine the data and masks generators to provide images and masks together
train_generator = zip(train_image_generator, train_mask_generator)
```

```python
# Define image and mask generators
image_datagen = ImageDataGenerator(
    rotation_range=20,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)
mask_datagen = ImageDataGenerator(
    rotation_range=20,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)

# Create generators for images and masks

image_generator = image_datagen.flow_from_directory(
    train_image_dir,
    target_size=(256, 256),
    batch_size=batch_size,
    class_mode=None,
    seed=42
)
mask_generator = mask_datagen.flow_from_directory(
    train_mask_dir,
    target_size=(256, 256),
    batch_size=batch_size,
    class_mode=None,
    seed=42
)

# Combine generators into a single generator that yields image and mask pairs
train_generator = zip(image_generator, mask_generator)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Calculate the total number of training samples
n_train_samples = len(train_image_generator.filenames)

# Calculate the steps per epoch
steps_per_epoch = n_train_samples // batch_size

model = Sequential()
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
epochs = 10
model.fit(train_generator, epochs=epochs, steps_per_epoch=steps_per_epoch)
model.save('/content/drive/MyDrive/biomed/testingMAsks/model_new_one.h5')
```

```python
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import array_to_img

model_path = '/content/drive/MyDrive/biomed/testingMAsks/model_new_one.h5'
image_path = '/content/drive/MyDrive/biomed/firstRun/train_image/Images Used/004.jpg'

# Load the model
model = load_model(model_path)

# Load and preprocess the image
image = load_img(image_path, target_size=(256, 256))
x = img_to_array(image)
x = np.expand_dims(x, axis=0)


# Predict the segmentation mask
y_pred = model.predict(x)[0]
y_pred = np.round(y_pred)
print(y_pred)

# Convert the mask to an image
mask = array_to_img(y_pred)

# Show the original image and predicted mask side by side
image.show()
mask.show()
```
